```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("first-test").master("local[*]").getOrCreate()
data = [(1, "Январь"),(2, "Февраль"),(3, "Март")]
df = spark.createDataFrame(data, ["индекс месяца","название"])
df.show()
spark.stop()
```
